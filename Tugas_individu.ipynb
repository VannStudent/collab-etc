```python
import pandas as pd
import matplotlib.pyplot as plt

# Daftar kolom untuk pie chart
pie_charts = [
    'Occupation',
    'BMI Category',
    'Gender',
    'Sleep Disorder',
    'Age',
    'Physical Activity Level',
    'Stress Level',
    'Quality of Sleep',
    'Smoking Status',
    'Alcohol Consumption'
]

# Hanya kolom yang tersedia
available_columns = [col for col in pie_charts if col in data.columns]

# Ubah kolom 'Age' jadi kategori jika perlu
if 'Age' in available_columns and pd.api.types.is_numeric_dtype(data['Age']):
    data['Age Group'] = pd.cut(
        data['Age'],
        bins=[0, 20, 30, 40, 50, 60, 100],
        labels=['<20', '21-30', '31-40', '41-50', '51-60', '60+']
    )
    available_columns = ['Age Group' if col == 'Age' else col for col in available_columns]

# Loop pie chart
for i, column in enumerate(available_columns[:10]):
    plt.figure(figsize=(8, 8))
    counts = data[column].value_counts()

    if len(counts) > 3:
        # Gunakan legenda agar label tidak tumpang tindih
        wedges, texts, autotexts = plt.pie(
            counts,
            labels=None,
            autopct='%1.1f%%',
            startangle=90,
            counterclock=False,
            textprops={'fontsize': 10}
        )
        plt.legend(wedges, counts.index, title=column, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    else:
        # Kalau kategorinya sedikit, tampilkan label langsung
        plt.pie(
            counts,
            labels=counts.index,
            autopct='%1.1f%%',
            startangle=90,
            counterclock=False,
            labeldistance=1.1,
            textprops={'fontsize': 10}
        )

    plt.title(f'Pie Chart {i+1}: Distribusi {column}')
    plt.tight_layout()
    plt.show()
```
